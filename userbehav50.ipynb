```python
user_counts = df.groupby('UserID').size().reset_index(name='Record_Count')
print(user_counts)
```

```python
user_counts.describe()
```

```python
user_counts_df = user_counts[user_counts['Record_Count'] >= 10]
user_counts_df.head()
```

```python
users_to_test = user_counts_df['UserID'].unique()[:100]
filtered_df = user_counts_df[user_counts_df['UserID'].isin(users_to_test)]
len(filtered_df)
```

```python
filtered_df = filtered_df.merge(df, on='UserID', how='left')
filtered_df.head()
```

```python
filtered_df.head()
```

```python
sample_path=r'C:\Users\marwa\Downloads\val_seq_sample.csv'
filtered_df.to_csv(sample_path,  index=False)
```
